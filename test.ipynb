```python
import pandas as pd
df= pd.read_csv('krx/201901_202201_ohlc.csv')
```

```python
df= df.drop(columns=['Unnamed: 0','index'])
```

```python
from datetime import datetime
df['Date']=[datetime.strptime(d,'%Y-%m-%d') for d in df.date]

T=list(set(df.티커.tolist()))
Date= sorted(list(set([datetime.strptime(d,'%Y-%m-%d') for d in df.date])))
M_Data=pd.DataFrame(columns=T,index=Date)
```

```python
d= df[['Date','티커','종가']]
```

```python
d=d.sort_values(['티커','Date'])
d=d.reset_index(drop=True)
d
```

```python
Test=pd.DataFrame(index=Date)
for j in M_Data.columns:
   test=d.loc[(d.티커 ==j)]
   test.index= test.Date
   if len(test) < len(M_Data):
      A=pd.concat([Test,test],axis=1)['종가'].values
   else:
      A=d.loc[(d.티커 ==j)]['종가'].values   
   M_Data.loc[:,j]=A
   
```

```python
M_Data.to_csv('2019_2022_close.csv')
```
